## Merge The Rest

```python
df_latlong_done=pd.read_csv(INTERIM_PATH+'20190901_sekolah_genap2019_data_latlong_all_part_one.csv')
del df_latlong_done['Unnamed: 0']
```

```python
data_list_part_two=[pd.read_csv(INTERIM_PATH_TMP+fn) 
                    for fn in os.listdir(INTERIM_PATH_TMP) if 'data_all_school' in fn]
df_latlong_done2=pd.concat(data_list_part_two)
del df_latlong_done2['level_0']
del df_latlong_done2['Unnamed: 0']
del df_latlong_done2['index']
df_latlong_done2=df_latlong_done2.reset_index()
del df_latlong_done2['index']
```

```python
df_latlong_done_all=pd.concat([df_latlong_done,df_latlong_done2])
df_latlong_done_all=df_latlong_done_all.reset_index()
del df_latlong_done_all['index']
```

```python
df_latlong_done_all.npsn=df_latlong_done_all.npsn.astype(str)
```

```python
## Get Existing npsn
all_fold_kabkot=[fold for fold in os.listdir(INTERIM_PATH2) if "." not in fold]

# kabkot=all_fold_kabkot[0]
kec_files=[]

for kabkot in all_fold_kabkot:
    df_all_files=[pd.read_csv(INTERIM_PATH2+kabkot+"/"+fkec) for fkec in os.listdir(INTERIM_PATH2+kabkot) if "ipynb" not in fkec]
    all_url_kec=pd.concat(df_all_files)
    all_url_kec=all_url_kec.reset_index()
    del all_url_kec['index']
    kec_files.append(all_url_kec)
    
school_urls=pd.concat(kec_files)
```

```python
school_urls.npsn=school_urls.npsn.astype(str)
```

```python
df_latlong_done_all.coordinate=df_latlong_done_all.coordinate.astype(str)
```

```python
def extract_lintang(x):
    try:
        content=x.split(',')
        if len(content)>2:
            content0=content[-2:]
        else:
            content0=content.copy()
            
        lin=float(content0[0].split(':')[1])
    except:
        lin=np.nan
    return lin

def extract_bujur(x):
    try:
        content=x.split(',')
        if len(content)>2:
            content0=content[-2:]
        else:
            content0=content.copy()
            
        buj=float(content0[1].split(':')[1]) 
    except:
        buj=np.nan
    return buj
```

```python
df_latlong_done_all['lintang']=df_latlong_done_all.coordinate.apply(lambda x: extract_lintang(x)).astype(float)
df_latlong_done_all['bujur']=df_latlong_done_all.coordinate.apply(lambda x: extract_bujur(x)).astype(float)
```
